```python
import matplotlib.pyplot as plt
```

```python
x1 = int(input())
x2 = int(input())

y1 = int(input())
y2 = int(input())
```

```python
dx = x2 - y1
dy = y2 - y1
```

```python
if abs(dx) > abs(dy):
    steps = abs(dx)
else:
    steps = abs(dy)
```

```python
xincrement = dx/steps
yincrement = dy/steps
```

```python
i = 0
```

```python
xcoordinates = []
ycoordinates = []

while i < steps:
    i += 1
    x1 = x1 + xincrement
    y1 = y1 + yincrement
    
    print("x1: ", x1, ", y1: ", y1)

    xcoordinates.append(x1)
    ycoordinates.append(y1)
```

```python
plt.plot(xcoordinates, ycoordinates)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('DDA Algorithm')
plt.show()
```
